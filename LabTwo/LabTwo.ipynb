<h4 style="font-family:sans-serif;">1. Створіть вектор розміром 11 зі значеннями від 0 до 5. Розрахуйте на
його основі вектор y (y = x^2).</h4>

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** (1/2)
```

<h4 style="font-family:sans-serif;">2. Побудова графіків.</h4>

plt.xlabel("X Label")<h4 style="font-family:sans-serif;">2.1 Функціональний метод. Для побудови базового графіку, що
відображає залежність х від у використовуйте наступну команду:</h4>

```python
plt.plot(x, y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```python
#subplot usage

plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y, x, "b")
```

<h4 style="font-family:sans-serif;">2.2. Об’єктно-орієнтований метод. Етапи створення графіку:
- ініціалізуйте створення графіку як об’єкта;
- додайте до нього поле, для відображення графіка.</h4>

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes = [fig.add_axes([0, 0, 1, 1]), fig.add_axes([0.5, 0.3, 0.4, 0.3])]

axes[0].plot(x, y, 'r')
axes[0].set_xlabel("X label")
axes[0].set_ylabel("Y label")
axes[0].set_title("Title")

axes[1].plot(x, y, 'b')
axes[1].set_xlabel("X label")
axes[1].set_ylabel("Y label")
axes[1].set_title("Title")
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4)

axes[0][0].plot(x,y)
axes[0][1].plot(y,x)
axes[0][2].plot(list(map(lambda el: el*2,x)),y)
axes[3][3].plot(list(map(lambda el: el*2,y)),x)
plt.tight_layout()
```

<h4>Додання легенди до графіку</h4>

```python
fig = plt.figure(figsize=(8,4), dpi=100)
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, x**2, label="x**2", color="red", lw=2, ls = '--')
axes.plot(x, x**4, label="x**3", color="green", marker='o',
          markersize=10, markerfacecolor= 'y')
axes.set_title("Title")
axes.legend()
```
