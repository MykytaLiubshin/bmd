<h3>Візуалізація даних за допомогою бібліотеки Seaborn</h3>

```python
import seaborn as sns
```

```python
%matplotlib inline
```

<h4>Завантаження даних з наборів sns</h4>

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head(10)
```

<h4><b>Візуалізація даних</b></h4>

<h4><b>Графіки для аналізу кількісних даних</b></h4>

```python
sns.distplot(diamonds['carat'], bins=20);
```

```python
sns.jointplot(x='carat', y='price', data=diamonds,  kind="reg");
```

```python
sns.pairplot(diamonds, hue="cut");
```

<h4><b>Графіки для аналізу категоріальних даних</b></h4>

```python
sns.barplot(x='color', y='price', data=diamonds);
```

```python
import numpy as np
sns.barplot(x='color', y='price', data=diamonds, estimator=np.mean);
```

```python
sns.barplot(x='color', y='price', data=diamonds, estimator=np.median);
```

```python
sns.boxplot(x='color', y='price', data=diamonds, palette='rainbow');
```

```python
sns.stripplot(x='color', y='price', data=diamonds);
```

<h4><b>Графіки для аналізу взаємозв’язку даних</b></h4>

```python
sns.heatmap(diamonds.corr())
```
